#### Step No. 4: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# The code below will:

# 1. Create a new column, “max_order,” into which I’ll place the results of the aggregation.
# CODE: ords_prods_merge['max_order'] =

# 2. Split the data into groups based on the “user_id” column.
# CODE: ords_prods_merge.groupby(['user_id'])

# 3. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# CODE: ['order_number'].transform(np.max)

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

#### Checking results

```python
ords_prods_merge[['user_id', 'order_number', 'max_order']].head(25)
```

#### Flag's filters

```python
# Users with more than 40 orders = 'Loyal customer'

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
# Users with 11 - 40 orders = 'Regular customer'

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Users with less than 11 orders = 'New customer'

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

#### Let's check frequency

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

#### Step No. 6: The team now wants to target different types of spenders in their marketing campaigns. 

```python
# This can be achieved by looking at the prices of the items people are buying. 
# Create a spending flag for each user based on the average price across all their orders using the following criteria:

# 1) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

# 2) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

#### Procedure:

```python
# As in Step No. 4, I'll first create a variable to store the mean of 'prices' grouped by 'user_id'

# This will give me a column (avg_spend_user) showing the average price paid per user.
```

```python
ords_prods_merge['avg_spend_user'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

#### Let's check the result:

```python
ords_prods_merge[['user_id', 'prices', 'avg_spend_user']].head(25)
```

#### Now, let's set up the filters as specified by the marketing team (mean lower than 10 = low spender | mean higher than or equal to 10 = high spender)

```python
ords_prods_merge.loc[ords_prods_merge['avg_spend_user'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_spend_user'] >= 10, 'spending_flag'] = 'High spender'
```

#### Frequency

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

#### Step No. 7: In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers.

```python
# Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 
# The criteria for the flag should be as follows:

# 1) If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# 2) If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# 3) If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

#### Almost identical procedure than that of steps 4 (max) and 6 (mean). This time, the value I'm looking for is the median.

```python
ords_prods_merge['ord_regularity_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

#### Checking the results under the column 'org_regularity_median'

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'ord_regularity_median']].head(25)
```

#### Setting up filters

```python
ords_prods_merge.loc[ords_prods_merge['ord_regularity_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['ord_regularity_median'] > 10) & (ords_prods_merge['ord_regularity_median'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['ord_regularity_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

#### Checking results

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'ord_regularity_median', 'frequency_flag']].head(25)
```

#### Frequency

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
